```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm
import time
%matplotlib inline

import tensorflow as tf
from sklearn.model_selection import train_test_split
from scipy.io import loadmat
```

### 1. Loading the dataset
Instead of loading the datasets from .csv file, it seems better for us to load them directly from .mat files, which could save us a lot of time and memory, especially when you would like to train your model on "extra" dataset. 

You can download the dataset in .mat format thought: http://ufldl.stanford.edu/housenumbers/

```python
def dense_to_one_hot(labels_dense, num_classes=10):
    """Convert class labels from scalars to one-hot vectors."""
    num_labels = labels_dense.shape[0]
    index_offset = np.arange(num_labels) * num_classes
    labels_one_hot = np.zeros((num_labels, num_classes))
    labels_one_hot.flat[index_offset + labels_dense.ravel()] = 1
    return labels_one_hot
```

```python
mat_train = loadmat("data/train_32x32.mat")
mat_train['X'] = mat_train['X'].transpose(3, 0, 1, 2)
X_train = mat_train['X']
X_train = 1.0 * X_train/255  # This line is vital for training!
y_train = mat_train['y'].reshape(-1, )
y_train[y_train == 10] = 0 

mat_test = loadmat('data/test_32x32.mat')
mat_test['X'] = mat_test['X'].transpose(3, 0, 1, 2)
X_test = mat_test['X']
X_test = 1.0*X_test/255
```

Split the training set

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=0)
```

#### Show me the image

```python
SAMPLE_IMAGE_INDEX = 0
plt.imshow(X_train[SAMPLE_IMAGE_INDEX])
print('y_train[{0}] => {1}'.format(SAMPLE_IMAGE_INDEX, y_train[0]))
```

### 2. Some useful functions, taken from
https://github.com/tensorflow/tensorflow/blob/master/tensorflow/examples/tutorials/mnist/mnist_deep.py

```python
def conv2d(x, W):
    """conv2d returns a 2d convolution layer with full stride."""
    return tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding='SAME')


def max_pool_2x2(x):
    """max_pool_2x2 downsamples a feature map by 2X."""
    return tf.nn.max_pool(x, ksize=[1, 2, 2, 1],
                        strides=[1, 2, 2, 1], padding='SAME')

def weight_variable(shape):
    """weight_variable generates a weight variable of a given shape."""
    initial = tf.truncated_normal(shape, stddev=0.1)
    return tf.Variable(initial)


def bias_variable(shape):
    """bias_variable generates a bias variable of a given shape."""
    initial = tf.constant(0.1, shape=shape)
    return tf.Variable(initial)
```

### 3. Build Tensorflow computation graph

```python
tf.reset_default_graph()

LEARNING_RATE = 1e-4
```

The placeholder is something that you are required to feed the graph when you training or test the model. Usually, we feed the graph with input, labels and some configuration. See https://www.tensorflow.org/api_docs/python/tf/placeholder for more details.

```python
# A placeholder for input
x = tf.placeholder(tf.float32, [None, 32, 32, 3], name='x') 
# A placeholder for labels
y_ = tf.placeholder(tf.float32, [None, 10], name='y_')
```

#### Now we setup the computational graph (the structure of our model)

The first convolutional layer's weight has a shape of [7, 7, 3, 32], which means the first layer will have 32 filters and the filters size of the first layer is 7 (height) by 7 (width) by 3 (depth). Following the conv layer is 2x2 max pooling.

```python
W_conv1 = weight_variable([3, 3, 3, 32])
b_conv1 = bias_variable([32])

# Conv-relu
h_conv1 = tf.nn.relu(conv2d(x, W_conv1) + b_conv1)
print("h_conv1: ")
print(h_conv1.shape)

layer1 = tf.reshape(h_conv1, [-1, 32, 32, 4, 8])
layer1 = tf.transpose(layer1, (0, 3, 1, 4, 2))
layer1 = tf.reshape(layer1, (-1, 32*4, 32*8)) 

# Pooling
h_pool1 = max_pool_2x2(h_conv1)
print("h_pool1: ")
print(h_pool1.shape)
```

#### The second conv-pooling layer

```python
W_conv2 = weight_variable([3, 3, 32, 64])
b_conv2 = bias_variable([64])

# Conv-relu
h_conv2 = tf.nn.relu(conv2d(h_pool1, W_conv2) + b_conv2)
print("h_conv2: ")
print(h_conv2.shape)

layer2 = tf.reshape(h_conv2, [-1, 16, 16, 4, 16])
layer2 = tf.transpose(layer2, (0, 3, 1, 4, 2))
layer2 = tf.reshape(layer2, (-1, 16*4, 16*16)) 

# Pooling
h_pool2 = max_pool_2x2(h_conv2)
print("h_pool2: ")
print(h_pool2.shape)
```

#### Affine-dropout-relu layer (fully connected layer)

```python
h_pool2_flat_shape = h_pool2.shape[1] * h_pool2.shape[2] * h_pool2.shape[3]
print(h_pool2_flat_shape)
```

```python
h_pool2_flat = tf.reshape(h_pool2, [-1, h_pool2_flat_shape])
print("h_pool2_flat: {}".format(h_pool2_flat.get_shape()))
```

```python
W_fc1 = weight_variable([h_pool2_flat_shape, 1024])
b_fc1 = bias_variable([1024])

h_fc1 = tf.nn.relu(tf.matmul(h_pool2_flat, W_fc1) + b_fc1)

# A dropout layer, the keep probability will be feeded to a placeholder
keep_prob = tf.placeholder(tf.float32)
h_fc1_dropout = tf.nn.dropout(h_fc1, keep_prob)
```

#### Affine layer (softmax layer)

```python
W_fc2 = weight_variable([1024, 10])
b_fc2 = bias_variable([10])

y = tf.matmul(h_fc1_dropout, W_fc2) + b_fc2
y_op = tf.nn.softmax(y) 
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_, logits=y))
#cross_entropy = -tf.reduce_sum(y_* tf.log(y_op))
```

```python
train_op = tf.train.AdamOptimizer(LEARNING_RATE).minimize(cross_entropy)
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
predict = tf.argmax(y,1)
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
epoch_index = 0
batch_index = 0
DISPLAY_STEP = 50
BATCH_SIZE = 512
NUM_EPOCH = 50
NUM_STEP = int((len(X_train) / BATCH_SIZE) * NUM_EPOCH)

DISPLAY_INDEX = 0

train_acc = []
train_loss = []
train_range = []
val_acc = []
val_loss = []
val_range = []
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
print("Training process will be ended after {} steps".format(NUM_STEP))
print("Initialized")
for step in range(NUM_STEP):
    offset = (batch_index * BATCH_SIZE) 
    X_batch = X_train[offset:(offset+BATCH_SIZE), :, :, :]
    y_batch = y_train[offset:(offset+BATCH_SIZE)]
    y_batch = dense_to_one_hot(y_batch)
    batch_index += 1
    # Shuffle the dataset and output validation accuracy
    if offset + BATCH_SIZE >= y_train.shape[0]:
        #print(offset + BATCH_SIZE)
        batch_index = 0
        '''perm = np.arange(y_train.shape[0])
        np.random.shuffle(perm)
        X_train = X_train[perm]
        y_train = y_train[perm]'''
        val_index = np.random.choice(X_val.shape[0], BATCH_SIZE)
        X_batch = X_val[val_index]
        y_batch = dense_to_one_hot(y_val[val_index])
        feed_dict = {x:X_batch, 
                 y_:y_batch, 
                 keep_prob:1.0}
        acc, loss = sess.run([accuracy, cross_entropy], feed_dict=feed_dict)
        val_acc.append(acc)
        val_loss.append(loss)
        val_range.append(epoch_index)
        print("Epoch: {0}, validation accuruay:{1:.4%}, validation loss: {2:.5f}"
              .format(epoch_index, acc, loss))
        epoch_index += 1
    feed_dict = {x:X_batch, 
                 y_:y_batch, 
                 keep_prob:0.5}
    _, acc, loss = sess.run([train_op, accuracy, cross_entropy], feed_dict=feed_dict)
    train_acc.append(acc)
    train_range.append(step)
    if step % DISPLAY_STEP == 0:
        print("Step: {0}, training accuracy: {1:.4%}, training loss: {2:.5f}"
              .format(step, acc, loss))
```

```python
plt.plot(train_range, train_acc, '-b', label='training accuracy')
plt.legend(loc='lower right', frameon=False)
plt.ylabel('accuracy')
plt.xlabel('step')
plt.show()
```

```python
plt.plot(val_range, val_acc, '-g', label='validation accuracy')
plt.legend(loc='lower right', frameon=False)
plt.ylabel('accuracy')
plt.xlabel('step')
plt.show()
```

### Test our model

```python
TEST_SAMPLE = 1
test_sample = X_test[TEST_SAMPLE:TEST_SAMPLE+1]
plt.imshow(X_test[TEST_SAMPLE])
```

```python
pred = sess.run(predict, feed_dict={
        x: test_sample,
        keep_prob: 1.0
    })
print("Prediction of test sample: {}".format(pred))
```

### Create the submission file
This submission should have an accuracy around 80% on pubilc leaderboard

```python
predicted_lables = np.zeros(X_test.shape[0])
for i in range(0, X_test.shape[0] // BATCH_SIZE):
    predicted_lables[i*BATCH_SIZE:(i+1)*BATCH_SIZE] = \
        sess.run(predict, feed_dict={x:X_test[i*BATCH_SIZE:(i+1)*BATCH_SIZE], keep_prob:1.0})
```

```python
predicted_lables[predicted_lables==0] = 10
```

```python
res_output = pd.DataFrame({
    "ImageId": range(len(predicted_lables)),
    "label": predicted_lables.astype(int)
})
```

```python
filename = "submission_2layerCNN.csv"
res_output.to_csv(filename, index=False)

print(filename)
print("Submission file created!")
```
